### Data visualization using matplotlib and pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#converting excel to csv file
data_xls = pd.read_excel('spike_history.xlsx', dtype=str, index_col=None)
data_xls.to_csv('spike_csvfile.csv', encoding='utf-8', index=False)
```

```python
data_sp=pd.read_csv('spike_csvfile.csv')
```

```python
#dataset of VALORANT match records
data_sp
```

```python
plt.style.available
```

```python
x_ticks=data_sp.duration # match duration column
x_list=x_ticks.tolist()
x_list.sort()
print(x_list)
```

```python
# duration VS avg_score
plt.figure(figsize=(10,6))

plt.title("Relation b\w duration and avg combat score",fontdict={'fontsize': 22})
plt.plot(x_list,data_sp['combat_score'],'r.--')
print(type(x_list[::3]))

plt.xticks(x_list[::3])
plt.show()
```

```python
#pie chart showing agents used in percentages

labels=['raze','phonix','viper','sage','sova']

plt.pie(data_sp.Agent_used.value_counts(),explode=[0,0.1,0,0,0],labels=labels,radius=1,autopct='%0.2f%%',shadow=True,wedgeprops={'width':1})
plt.title("Agents Used",fontdict={'fontsize': 22})
plt.show()
```

```python
plt.hist(data_sp['kills'])
```

```python


wins=data_sp[data_sp['outcome']=='w'].count()[0]
losts=data_sp[data_sp['outcome']=='l'].count()[0]
```

```python
plt.pie([wins,losts],labels=['won','lost'],autopct='%0.2f%%',shadow=True)
plt.title("W/L percentage",fontdict={'fontsize': 22})
plt.show()
```

```python
#maps played
haven=data_sp.loc[data_sp['map']=='haven'].count()[0]
ascent=data_sp.loc[data_sp['map']=='ascent'].count()[0]
bind=data_sp.loc[data_sp['map']=='bind'].count()[0]
split=data_sp.loc[data_sp['map']=='split'].count()[0]
maps=[haven,ascent,bind,split]
maps
```

```python
plt.pie(maps,labels=['haven','ascent','bind','split'],explode=[0,0.1,0,0],autopct='%0.2f%%',shadow=True,radius=1)
plt.title("Playing percentage in all maps")
plt.show()
```

```python
#avg_score in every map 

h=data_sp.loc[data_sp['map']=='haven']['combat_score']
a=data_sp.loc[data_sp['map']=='ascent']['combat_score']
b=data_sp.loc[data_sp['map']=='bind']['combat_score']
s=data_sp.loc[data_sp['map']=='split']['combat_score']
plt.figure(figsize=(10,8))
labels=['haven','ascent','bind','split']
boxprops = dict(linewidth=2)
plt.boxplot([h,a,b,s],labels=labels,boxprops=boxprops)
plt.ylabel('Avg combat score')
plt.title('Comparison of maps according to combat score',fontdict={'fontsize': 18})
plt.show()
```

```python
h=data_sp[(data_sp['map']=='haven')]
hw=h.loc[h.outcome=='w'].count()[0]

a=data_sp[(data_sp['map']=='ascent')]
aw=a.loc[a.outcome=='w'].count()[0]

b=data_sp[(data_sp['map']=='bind')]
bw=b.loc[b.outcome=='w'].count()[0]

s=data_sp[(data_sp['map']=='split')]
sw=s.loc[s.outcome=='w'].count()[0]

map_w=[hw,aw,bw,sw]
plt.figure(figsize=(5,7))
plt.title('Winning caparison for each map',fontdict={'fontsize': 22})
plt.pie(map_w,labels=['haven','ascent','bind','split'],explode=[0,0.1,0,0],autopct='%0.2f%%',shadow=True,radius=1)
#plt.legend()
plt.show()
```

```python
h=data_sp.loc[data_sp['Agent_used']=='raze']['kills']
a=data_sp.loc[data_sp['Agent_used']=='phonix']['kills']
b=data_sp.loc[data_sp['Agent_used']=='viper']['kills']
s=data_sp.loc[data_sp['Agent_used']=='sage']['kills']
so=data_sp.loc[data_sp['Agent_used']=='sova']['kills']
plt.figure(figsize=(10,8))
labels=['raze','phonix','viper','sage','sova']
boxprops = dict(linewidth=2)
plt.boxplot([h,a,b,s,so],labels=labels,boxprops=boxprops)
plt.ylabel('Kills per match')
plt.title('Comparison of kills by every agents used',fontdict={'fontsize': 18})
plt.show()
```

###  data visualization using seaborn

```python
import numpy as np 
import seaborn as sns
%matplotlib inline
```

```python
data_sp.head(5)
```

```python
duration_l=[]
for d in data_sp.duration:
    n=d[0:2]+'.'+d[3:5]
    duration_l.append(float(n))
    
duration_l  
```

```python
ax = sns.lineplot(x='kills', y='combat_score', hue='outcome' ,data = data_sp)
plt.title("Kills Vs Avg score")
```

```python
data_sp['rd_w']
```

```python
sns.set(rc={'figure.figsize':(8,8)})
ax = sns.lineplot(x=duration_l, y=abs(data_sp['rd_w']-data_sp['rd_lost']), hue='outcome' ,data = data_sp)
plt.xlabel("Duration of a match(mins)")
plt.ylabel("difference in rounds")
```

```python
sns.set(rc={'figure.figsize':(5,5)})
plt.title("Duration  Vs KDA")
ax = sns.scatterplot(x=duration_l, y=((data_sp['kills']+data_sp['Assists'])/data_sp['deaths']),size = 'combat_score' ,data = data_sp,sizes=(50,200))
```

```python
x = data_sp['combat_score'].values

sns.distplot(x, color = 'blue');
# Calculating the mean
mean = data_sp['combat_score'].mean()

#ploting the mean
plt.axvline(mean, 0,1, color = 'red')
```

```python
x = ((data_sp['kills']+data_sp['Assists'])/data_sp['deaths']).values

sns.distplot(x, color = 'blue');
# Calculating the mean
mean = ((data_sp['kills']+data_sp['Assists'])/data_sp['deaths']).mean()

#ploting the mean
plt.axvline(mean, 0,1, color = 'red')
```
